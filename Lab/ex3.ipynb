```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
df = pd.read_csv('./Data/advertising.csv')
```

```python


sales = df["Sales"].dropna()

mu = sales.mean()
sigma = sales.std()

s = 10
z = (s - mu) / sigma

xmin, xmax = sales.min(), sales.max()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)

plt.hist(sales, bins=20, density=True)
plt.plot(x, p)

plt.axvline(mu)
plt.axvline(mu - sigma)
plt.axvline(mu + sigma)
plt.axvline(s)

plt.title("Sales Distribution with Normal Curve")
plt.xlabel("Sales")
plt.ylabel("Density")
plt.show()

```
