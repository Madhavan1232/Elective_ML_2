```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_id = "1SaC087F3jrW5rPkcUOtVeQ2vraxR30NL"
url = f"https://drive.google.com/uc?id={file_id}"

df = pd.read_csv(url)
n = df['delivery_time'].size
sample = df['delivery_time'].sample(n=n, replace=False)

mean  = sample.mean()
_std =  sample.std()

print(f"Mean Delivery Time: {mean}")
print(f"Standard Deviation of Delivery Time: {_std}")
```

```python

file_id = "1wzCEsY8QU-L0kPwbStxz-J1hvL24MBJD"
url = f"https://drive.google.com/uc?id={file_id}"

df = pd.read_csv(url)
```

```python
sample_items = {
    'Electronics' : 4,
    'Grocery' : 2,
    'Clothing' : 4,
}

sample_df = df.groupby('Category' , group_keys=False).apply(
    lambda x: x.sample(sample_items[x.name] , replace=False).assign(Category = x.name),
    include_groups = True
).reset_index(drop=True)
sample_df
```

```python
population = np.random.exponential(scale=50, size=10000)

plt.hist(population, bins=50, alpha=0.7, edgecolor='black')
plt.show()
sample_means = []
for i in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(sample.mean())

plt.hist(sample_means, bins=30, alpha=0.7, edgecolor='black')
plt.show()
```
