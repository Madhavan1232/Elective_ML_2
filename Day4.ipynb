```python
from scipy.stats import chisquare


observed_values = [20 , 30, 25, 25]
expected_values = [25, 25, 25, 25]

chi2_stat, p_value = chisquare(observed_values, expected_values)
print(f"Chi-squared Statistic: {chi2_stat}, P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("We reject the idea that students like all flavors equally.")
else:
    print("We do not reject the idea that students like all flavors equally.")
```
