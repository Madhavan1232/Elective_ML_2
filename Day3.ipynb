```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data: battery life in hours for 100 devices
sample_data = np.random.normal(loc=11.8, scale=0.5, size=100)

# Known population mean (claimed by manufacturer)
population_mean = 12

# Perform one-sample Z-test
z_stat, p_value = ztest(sample_data, value=population_mean)

print("One-Sample Z-Test")
print("Z-Statistic:", z_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis: The average battery life differs from 12 hours.")
else:
    print("Fail to reject null hypothesis: No significant difference from 12 hours.")
```
