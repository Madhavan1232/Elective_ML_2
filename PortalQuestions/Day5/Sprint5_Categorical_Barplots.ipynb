```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load & basic cleanup
csv_path = '../Data/ML374_S5_Practice_Income_Data.csv'
df = pd.read_csv(csv_path)
df.columns = df.columns.str.strip()
df.replace('?', np.nan, inplace=True)

# exact columns and layout to match the provided image
cols = ['WorkClass', 'Education', 'Occupation', 'Relationship', 'Gender', 'Native_Country']
n_cols = 3
n_rows = 2
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 6))
axes_flat = axes.flatten()

for i, col in enumerate(cols):
    ax = axes_flat[i]
    counts = df[col].fillna('Missing').value_counts()
    ax.bar(counts.index.astype(str), counts.values, color='#1f77b4', edgecolor='black')
    ax.set_title(col)
    ax.set_ylabel('')
    # rotate and tighten labels to mirror the example image
    ax.tick_params(axis='x', rotation=90)
    ax.set_xticklabels(counts.index.astype(str), fontsize=7)

# remove any unused axes (if any)
total_axes = n_rows * n_cols
for j in range(len(cols), total_axes):
    try:
        fig.delaxes(axes_flat[j])
    except Exception:
        pass

plt.tight_layout()
plt.show()
```
