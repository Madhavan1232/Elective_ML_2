```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

dataset = make_blobs(n_samples=5000, centers=5, cluster_std=0.60, random_state=42)
x, y = dataset
model = KMeans(n_clusters=5)
model.fit(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = 'orange' , s = 1 )
```

```python
centroids = model.cluster_centers_
plt.scatter(x[:, 0], x[:, 1], c = 'orange' , s = 1)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50, marker='X' , alpha=1)
plt.show()
```

```python
centroids = model.cluster_centers_
plt.scatter(x[:, 0], x[:, 1], c = 'orange' , s = 1)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50, marker='X' , alpha=1)
plt.scatter(3.0 , 1.5 , marker='*' , c='blue' , s=200 )
plt.show()
```
