```python
import numpy as np, pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
np.random.seed(7)
X = np.vstack([np.random.normal(0,1,(50,2)), np.random.normal(3,1,(50,2))])
y = np.array([0]*50 + [1]*50)
knn = KNeighborsClassifier(n_neighbors=5).fit(X,y)
pred = knn.predict(X)
print('KNN training accuracy=', accuracy_score(y,pred))
data = np.vstack([np.random.normal([0,0],0.7,(100,2)), np.random.normal([4,4],0.7,(100,2))])
inertia = []
for k in range(1,6):
    inertia.append(KMeans(n_clusters=k, random_state=0).fit(data).inertia_)
plt.plot(range(1,6), inertia, '-o'); plt.title('Elbow method')
km = KMeans(n_clusters=2, random_state=0).fit(data)
plt.figure(); plt.scatter(data[:,0], data[:,1], c=km.labels_, cmap='viridis', s=10); plt.title('KMeans clusters')
```
