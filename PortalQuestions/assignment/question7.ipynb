```python
import numpy as np, pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
np.random.seed(6)
X1 = np.random.normal(0,1,300)
X2 = np.random.normal(1,1,300)
logit_prob = 1/(1+np.exp(-(0.8*X1 + 0.5*X2 - 0.2)))
y = (logit_prob > 0.5).astype(int)
X = np.column_stack([X1,X2])
clf = LogisticRegression().fit(X,y)
pred_prob = clf.predict_proba(X)[:,1]
pred = clf.predict(X)
print('Confusion matrix:\n', confusion_matrix(y,pred))
print('AUC=', roc_auc_score(y,pred_prob))
fpr,tpr,_ = roc_curve(y,pred_prob)
plt.plot(fpr,tpr); plt.plot([0,1],[0,1],'k--'); plt.title('ROC')
```
