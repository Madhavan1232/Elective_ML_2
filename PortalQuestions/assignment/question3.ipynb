```python
import numpy as np, pandas as pd
from scipy import stats
np.random.seed(2)
cats = np.random.choice(['A','B','C'], size=300, p=[0.5,0.3,0.2])
obs = pd.Series(cats).value_counts().sort_index()
expected = np.array([0.5,0.3,0.2])*len(cats)
chi2, p = stats.chisquare(f_obs=obs, f_exp=expected)
print('Chi-square GOF:', chi2, p)
gender = np.random.choice(['M','F'], size=300)
purchased = np.random.choice([0,1], size=300, p=[0.6,0.4])
ct = pd.crosstab(gender, purchased)
chi2i, pi, dof, ex = stats.chi2_contingency(ct)
print('Chi-square independence:', chi2i, pi, 'dof', dof)
```
