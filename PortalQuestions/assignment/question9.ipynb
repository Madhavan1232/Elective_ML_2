```python
import numpy as np
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
np.random.seed(8)
X = np.vstack([np.random.normal(0,1,(100,2)), np.random.normal(2,1,(100,2))])
y = np.array([0]*100 + [1]*100)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
log = LogisticRegression()
knn = KNeighborsClassifier(n_neighbors=5)
print('Logistic CV accuracy mean=', cross_val_score(log,X,y,cv=skf).mean())
print('KNN CV accuracy mean=', cross_val_score(knn,X,y,cv=skf).mean())
log.fit(X,y); knn.fit(X,y)
for m,name in [(log,'Logistic'), (knn,'KNN')]:
    pred = m.predict(X)
    print(name, 'acc', accuracy_score(y,pred), 'precision', precision_score(y,pred), 'recall', recall_score(y,pred), 'f1', f1_score(y,pred))
```
