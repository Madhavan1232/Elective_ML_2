```python
import numpy as np, pandas as pd
from scipy import stats
np.random.seed(1)
groupA = np.random.normal(55, 10, size=100)  
groupB = np.random.normal(50, 12, size=120)  
tstat, pval = stats.ttest_1samp(groupA, 50)
print('One-sample t-test: t=', tstat, 'p=', pval)
t2, p2 = stats.ttest_ind(groupA, groupB, equal_var=False)
print('Two-sample t-test (Welch): t=', t2, 'p=', p2)
import math
mu0 = 50
xbar = groupA.mean(); se = groupA.std(ddof=1)/math.sqrt(len(groupA))
z = (xbar-mu0)/se
p_z = 2*(1-stats.norm.cdf(abs(z)))
print('One-sample Z approx: z=', z, 'p=', p_z)
```
