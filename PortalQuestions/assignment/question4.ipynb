```python
import numpy as np, pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
np.random.seed(3)
df = pd.DataFrame({'age': np.random.choice([25,35,45,None], size=100),
                   'income': np.append(np.random.normal(60000,15000,99), [1e6]),
                   'gender': np.random.choice(['M','F'], size=100)})
df['age_fill_mean'] = df['age'].fillna(df['age'].mean())
imputer = KNNImputer(n_neighbors=3)
df[['age_imputed']] = imputer.fit_transform(df[['age']])
Q1 = df['income'].quantile(0.25); Q3 = df['income'].quantile(0.75); IQR = Q3-Q1
outliers = df[(df['income'] < Q1 - 1.5*IQR) | (df['income'] > Q3 + 1.5*IQR)]
print('Outliers found:', len(outliers))
df['gender_le'] = LabelEncoder().fit_transform(df['gender'])
df = pd.concat([df, pd.get_dummies(df['gender'], prefix='g')], axis=1)
df['income_scaled'] = StandardScaler().fit_transform(df[['income']])
train, test = train_test_split(df, test_size=0.2, random_state=4)
print('Train/test sizes:', len(train), len(test))
```
