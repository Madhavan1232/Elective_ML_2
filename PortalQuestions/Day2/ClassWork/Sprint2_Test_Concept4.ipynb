```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

df = pd.read_excel("../../Data/ML374_S2_Height_Weight_Data_Concept.xlsx")
```

```python
population = df['Height_cm'].values
sample_size = 50
num_samples = 1000

sample_means = []
for _ in range(num_samples):
    sample = random.choices(population, k=sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

sns.histplot(sample_means, kde=True , color = 'lightcoral' ,bins=15) 
plt.title("Sampling Distribution of the Means (CLT in Action)")
plt.xlabel("Sample Means of Height (Cm) (n = 50)")
```
