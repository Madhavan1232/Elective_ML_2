```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("../Data/ML374_S2_Height_Weight_Data_Concept.xlsx")
```

```python
sns.histplot(data=df, x='Height_cm', kde=True , color = 'skyblue')
plt.title("Population")
```

```python
sample_df = df.sample(n=50, random_state=42)
sns.histplot(data=sample_df, x='Height_cm', kde=True, color='lightcoral')
plt.title("Sample ")
```

```python
sample_df_5 = df.sample(n=5 , random_state=42)
sns.histplot(data=sample_df_5 , x = 'Height_cm', kde=True, color='skyblue' , bins = 4)
plt.title("Sample of 5")
```
