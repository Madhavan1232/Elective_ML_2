```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("c:/Users/madha/PycharmProjects/Elective_ML_2/PortalQuestions/Data/ML374_S9_Practice_Diabetes_Data.csv")
df.head()
```

```python
# Basic structure and missing/zero inspection
print('Shape:', df.shape)
print('\nColumns:', list(df.columns))
print('\nCounts of zeros (may indicate missing values):')
print((df == 0).sum())

print('\nInfo:')
df.info()
print('\nDescriptive statistics:')
df.describe()
```

```python
features = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']

df_plot = df.copy()
df_plot[features] = df_plot[features].replace(0, pd.NA)

sns.set(style='whitegrid')
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()
for i, feat in enumerate(features):
    ax = axes[i]
    sns.histplot(df_plot[feat].dropna(), bins=30, kde=True, ax=ax, color='C0')
    ax.set_title(feat)
    ax.set_xlabel('')

axes[-1].set_visible(False)
plt.tight_layout()
plt.show()
```
